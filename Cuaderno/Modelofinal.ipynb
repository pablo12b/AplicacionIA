```python
# === 1. INSTALAR DEPENDENCIAS NECESARIAS ===
!pip install -q rembg onnxruntime
!pip install -q tensorflow matplotlib scikit-learn opencv-python

# === 2. DESCARGAR Y PREPARAR DATASET ===
!wget https://raw.githubusercontent.com/marcovvvv/dataset-clasificador/main/dataset-main.zip
!unzip -qo dataset-main.zip

# === 3. IMPORTAR LIBRER√çAS ===
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from PIL import Image
from rembg import remove
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.utils.class_weight import compute_class_weight

# === 4. CONFIGURACI√ìN ===
IMG_SIZE = (224, 224)  # Tama√±o compatible con MobileNetV2
BATCH_SIZE = 32
AUTOTUNE = tf.data.AUTOTUNE
EPOCHS_PRETRAIN = 30
EPOCHS_FINETUNE = 20
LR_PRETRAIN = 0.0001
LR_FINETUNE = 0.00001

# === 5. PROCESAMIENTO CON SEGMENTACI√ìN (ELIMINACI√ìN DE FONDO) ===
def procesar_imagenes_con_segmentacion():
    os.makedirs("dataset-segmentado", exist_ok=True)

    # Verificar existencia del directorio
    if not os.path.exists('dataset-main/images'):
        raise Exception("No se encontr√≥ el directorio 'dataset-main/images'")

    for img_name in os.listdir("dataset-main/images"):
        if img_name.lower().endswith((".jpg", ".jpeg", ".png")):
            try:
                img_path = os.path.join("dataset-main/images", img_name)

                # Usar OpenCV para cargar la imagen
                img = cv2.imread(img_path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Convertir a PIL para la segmentaci√≥n
                img_pil = Image.fromarray(img)

                # Segmentaci√≥n con rembg (eliminar fondo)
                img_segmentada = remove(img_pil)

                # Convertir a array numpy para procesamiento
                img_segmentada_np = np.array(img_segmentada)

                # Crear m√°scara binaria del objeto segmentado
                if img_segmentada_np.shape[2] == 4:  # Si tiene canal alpha
                    mask = img_segmentada_np[:,:,3] > 0
                    img_segmentada_np = img_segmentada_np[:,:,:3]
                else:
                    mask = np.all(img_segmentada_np > [0, 0, 0], axis=-1)

                # Aplicar m√°scara para fondo blanco
                img_final = np.ones_like(img_segmentada_np) * 255
                img_final[mask] = img_segmentada_np[mask]

                # Guardar imagen segmentada
                output_path = os.path.join("dataset-segmentado", img_name)
                cv2.imwrite(output_path, cv2.cvtColor(img_final, cv2.COLOR_RGB2BGR))

            except Exception as e:
                print(f"Error procesando {img_name}: {str(e)}")

procesar_imagenes_con_segmentacion()

# === 6. PREPARACI√ìN DE DATOS ===
labels_df = pd.read_csv("dataset-main/labels.csv")
labels_df['filename'] = labels_df['filename'].str.replace('img13jpg', 'img13.jpg').str.strip()

class_names = sorted(labels_df['labels'].unique())
class_to_index = {name: idx for idx, name in enumerate(class_names)}
labels_df['label_idx'] = labels_df['labels'].map(class_to_index)

print("\n=== INFORMACI√ìN DEL DATASET ===")
print(f"Total clases: {len(class_names)}")
print(f"Muestras totales: {len(labels_df)}")
print("\nDistribuci√≥n de clases:")
print(labels_df['labels'].value_counts())

# === 7. PREPROCESAMIENTO Y AUMENTACI√ìN DE DATOS ===
def preprocess_image(image):
    # Convert tensor to numpy array for OpenCV
    image_np = image.numpy().astype(np.uint8)
    # Mejorar contraste con CLAHE
    image_lab = cv2.cvtColor(image_np, cv2.COLOR_RGB2LAB)
    l, a, b = cv2.split(image_lab)
    clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
    l = clahe.apply(l)
    image_lab = cv2.merge((l,a,b))
    image_rgb = cv2.cvtColor(image_lab, cv2.COLOR_LAB2RGB)
    # Convert back to tensor
    return tf.convert_to_tensor(image_rgb, dtype=tf.float32)

def decode_and_augment(filename, label):
    img_path = tf.strings.join(["dataset-segmentado/", filename])
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img.set_shape((*IMG_SIZE, 3)) # Set shape

    # Aplicar preprocesamiento
    img = tf.py_function(preprocess_image, [img], tf.float32)
    img.set_shape((*IMG_SIZE, 3)) # Set shape after py_function

    # Aumentaci√≥n de datos
    img = tf.image.random_flip_left_right(img)
    img = tf.image.random_brightness(img, max_delta=0.1)
    img = tf.image.random_contrast(img, lower=0.9, upper=1.1)

    img = img / 255.0
    return img, label

def decode_image(filename, label):
    img_path = tf.strings.join(["dataset-segmentado/", filename])
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img.set_shape((*IMG_SIZE, 3)) # Set shape
    img = tf.py_function(preprocess_image, [img], tf.float32)
    img.set_shape((*IMG_SIZE, 3)) # Set shape after py_function
    img = img / 255.0
    return img, label

# === 8. PREPARAR DATASETS ===
X_train, X_val, y_train, y_val = train_test_split(
    labels_df['filename'].values,
    labels_df['label_idx'].values,
    test_size=0.2,
    random_state=42
)

train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)) \
    .map(decode_and_augment, num_parallel_calls=AUTOTUNE) \
    .shuffle(1000) \
    .batch(BATCH_SIZE) \
    .prefetch(AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((X_val, y_val)) \
    .map(decode_image, num_parallel_calls=AUTOTUNE) \
    .batch(BATCH_SIZE) \
    .prefetch(AUTOTUNE)

# === 9. PESOS DE CLASE PARA BALANCEO ===
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(labels_df['label_idx']),
    y=labels_df['label_idx']
)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}

# === 10. CONSTRUIR MODELO CON MOBILENETV2 ===
def build_model():
    # Data augmentation
    data_augmentation = tf.keras.Sequential([
        layers.RandomFlip("horizontal"),
        layers.RandomRotation(0.1),
        layers.RandomZoom(0.1),
    ])

    # Base MobileNetV2
    base_model = MobileNetV2(
        input_shape=(*IMG_SIZE, 3),
        include_top=False,
        weights="imagenet",
        alpha=1.0  # Versi√≥n completa
    )
    base_model.trainable = False

    # Construcci√≥n del modelo
    inputs = tf.keras.Input(shape=(*IMG_SIZE, 3))
    x = data_augmentation(inputs)
    x = base_model(x, training=False)
    x = layers.GlobalAveragePooling2D()(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.5)(x)

    x = layers.Dense(256, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(len(class_names), activation='softmax')(x)

    return tf.keras.Model(inputs, outputs)

model = build_model()

# Optimizador
optimizer = optimizers.Adam(
    learning_rate=LR_PRETRAIN,
    beta_1=0.9,
    beta_2=0.999
)

model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

# Callbacks
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=10,
        restore_best_weights=True
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=5,
        min_lr=1e-7
    ),
    ModelCheckpoint(
        'best_model.h5',
        monitor='val_accuracy',
        save_best_only=True
    )
]

# === 11. ENTRENAMIENTO ===
print("\nüî• Entrenamiento inicial")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_PRETRAIN,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# === 12. FINE-TUNING ===
print("\n Fine-tuning")
base_model = model.layers[2]
base_model.trainable = True

# Descongelar solo las √∫ltimas capas
for layer in base_model.layers[:100]:
    layer.trainable = False

model.compile(
    optimizer=optimizers.Adam(LR_FINETUNE),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_FINETUNE,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# === 13. GUARDAR MODELO ===
model.save("product_classifier.keras")
print("\n Modelo guardado como 'product_classifier.keras'")

# === 14. EVALUACI√ìN ===
def evaluate_model(model, dataset):
    results = model.evaluate(dataset, verbose=0)
    print(f"\nüîπ Loss: {results[0]:.4f}")
    print(f"üîπ Accuracy: {results[1]:.4f}")

    y_true, y_pred = [], []
    for images, labels in dataset:
        preds = model.predict(images, verbose=0)
        y_true.extend(labels.numpy())
        y_pred.extend(np.argmax(preds, axis=1))

    # Get unique labels present in the validation set
    unique_labels = np.unique(y_true + y_pred)
    target_names_subset = [class_names[i] for i in unique_labels]

    print("\n Reporte de clasificaci√≥n:")
    print(classification_report(y_true, y_pred, target_names=target_names_subset, labels=unique_labels, zero_division=0))

    # Matriz de confusi√≥n
    cm = confusion_matrix(y_true, y_pred, labels=unique_labels)
    plt.figure(figsize=(12, 10))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=target_names_subset)
    disp.plot(xticks_rotation=90, values_format='d')
    plt.title("Matriz de Confusi√≥n")
    plt.tight_layout()
    plt.show()

evaluate_model(model, val_ds)

# === 15. VISUALIZACI√ìN DE RESULTADOS ===
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy durante el Entrenamiento')
    plt.xlabel('√âpoca')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss durante el Entrenamiento')
    plt.xlabel('√âpoca')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

print("\n Visualizaci√≥n del progreso de entrenamiento:")
plot_training_history(history)
if 'history_fine' in locals():
    plot_training_history(history_fine)
```

```python
# === 1. INSTALAR DEPENDENCIAS ===
!pip install -q rembg onnxruntime
!pip install -q tensorflow matplotlib scikit-learn opencv-python imbalanced-learn
!pip install -q albumentations  # Para aumentaci√≥n de datos avanzada

# === 2. DESCARGAR DATASET ===
!wget https://raw.githubusercontent.com/marcovvvv/dataset-clasificador/main/dataset-main.zip
!unzip -qo dataset-main.zip

# === 3. IMPORTAR LIBRER√çAS ===
import tensorflow as tf
from tensorflow.keras import layers, models, optimizers, regularizers
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2
from PIL import Image
from rembg import remove
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
from imblearn.over_sampling import RandomOverSampler
from sklearn.utils.class_weight import compute_class_weight
import albumentations as A

# === 4. CONFIGURACI√ìN ===
IMG_SIZE = (224, 224)
BATCH_SIZE = 16  # Reducido para mejor generalizaci√≥n
AUTOTUNE = tf.data.AUTOTUNE
EPOCHS_PRETRAIN = 50
EPOCHS_FINETUNE = 30
LR_PRETRAIN = 0.0001
LR_FINETUNE = 0.00001

# === 5. PROCESAMIENTO CON SEGMENTACI√ìN MEJORADA ===
def procesar_imagenes_con_segmentacion():
    os.makedirs("dataset-segmentado", exist_ok=True)

    if not os.path.exists('dataset-main/images'):
        raise Exception("No se encontr√≥ el directorio 'dataset-main/images'")

    for img_name in os.listdir("dataset-main/images"):
        if img_name.lower().endswith((".jpg", ".jpeg", ".png")):
            try:
                img_path = os.path.join("dataset-main/images", img_name)
                img = cv2.imread(img_path)
                img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

                # Segmentaci√≥n mejorada
                img_pil = Image.fromarray(img)
                img_segmentada = remove(img_pil)
                img_segmentada_np = np.array(img_segmentada)

                # Manejo de transparencia y fondo
                if img_segmentada_np.shape[2] == 4:
                    mask = img_segmentada_np[:,:,3] > 0
                    img_segmentada_np = img_segmentada_np[:,:,:3]
                else:
                    mask = np.all(img_segmentada_np > [10, 10, 10], axis=-1)

                # Fondo blanco con bordes difuminados
                img_final = np.ones_like(img_segmentada_np) * 255
                img_final[mask] = img_segmentada_np[mask]

                # Guardar con calidad
                output_path = os.path.join("dataset-segmentado", img_name)
                cv2.imwrite(output_path, cv2.cvtColor(img_final, cv2.COLOR_RGB2BGR),
                            [cv2.IMWRITE_JPEG_QUALITY, 100])

            except Exception as e:
                print(f"Error procesando {img_name}: {str(e)}")

procesar_imagenes_con_segmentacion()

# === 6. PREPARACI√ìN DE DATOS CON BALANCEO ===
labels_df = pd.read_csv("dataset-main/labels.csv")
labels_df['filename'] = labels_df['filename'].str.replace('img13jpg', 'img13.jpg').str.strip()

class_names = sorted(labels_df['labels'].unique())
class_to_index = {name: idx for idx, name in enumerate(class_names)}
labels_df['label_idx'] = labels_df['labels'].map(class_to_index)

print("\n=== INFORMACI√ìN DEL DATASET ===")
print(f"Total clases: {len(class_names)}")
print(f"Muestras totales: {len(labels_df)}")
print("\nDistribuci√≥n de clases:")
print(labels_df['labels'].value_counts())

# Balanceo con sobremuestreo
ros = RandomOverSampler(random_state=42)
X_resampled, y_resampled = ros.fit_resample(
    labels_df[['filename']],
    labels_df['label_idx']
)

# === 7. AUMENTACI√ìN DE DATOS AVANZADA ===
augmenter = A.Compose([
    A.HorizontalFlip(p=0.5),
    A.VerticalFlip(p=0.2),
    A.Rotate(limit=15, p=0.5),
    A.RandomBrightnessContrast(p=0.3),
    A.GaussianBlur(blur_limit=(3, 5), p=0.2),
    A.CLAHE(p=0.5),
])

def apply_augmentation(image):
    augmented = augmenter(image=image.numpy().astype(np.uint8))
    return tf.convert_to_tensor(augmented['image'], dtype=tf.float32)

def decode_and_augment(filename, label):
    img_path = tf.strings.join(["dataset-segmentado/", filename])
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)

    # Aplicar aumentaci√≥n
    img = tf.py_function(apply_augmentation, [img], tf.float32)
    img.set_shape((*IMG_SIZE, 3))

    img = img / 255.0
    return img, label

def decode_image(filename, label):
    img_path = tf.strings.join(["dataset-segmentado/", filename])
    img = tf.io.read_file(img_path)
    img = tf.image.decode_jpeg(img, channels=3)
    img = tf.image.resize(img, IMG_SIZE)
    img = img / 255.0
    return img, label

# === 8. PREPARAR DATASETS ===
X_train, X_val, y_train, y_val = train_test_split(
    X_resampled['filename'].values,
    y_resampled,
    test_size=0.15,  # Menos datos para validaci√≥n
    stratify=y_resampled,
    random_state=42
)

train_ds = tf.data.Dataset.from_tensor_slices((X_train, y_train)) \
    .map(decode_and_augment, num_parallel_calls=AUTOTUNE) \
    .shuffle(2000) \
    .batch(BATCH_SIZE) \
    .prefetch(AUTOTUNE)

val_ds = tf.data.Dataset.from_tensor_slices((X_val, y_val)) \
    .map(decode_image, num_parallel_calls=AUTOTUNE) \
    .batch(BATCH_SIZE) \
    .prefetch(AUTOTUNE)

# === 9. PESOS DE CLASE ===
class_weights = compute_class_weight(
    class_weight='balanced',
    classes=np.unique(y_resampled),
    y=y_resampled
)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}

# === 10. MODELO OPTIMIZADO ===
def build_model():
    # Base MobileNetV2 con fine-tuning
    base_model = MobileNetV2(
        input_shape=(*IMG_SIZE, 3),
        include_top=False,
        weights="imagenet",
        alpha=1.0,
        pooling='avg'  # GlobalAveragePooling directamente
    )

    # Congelar capas inicialmente
    base_model.trainable = False

    # Capas personalizadas
    inputs = tf.keras.Input(shape=(*IMG_SIZE, 3))
    x = base_model(inputs, training=False)
    x = layers.Dropout(0.5)(x)

    # Capas densas con regularizaci√≥n
    x = layers.Dense(512, activation='relu',
                    kernel_regularizer=regularizers.l2(0.001))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)

    x = layers.Dense(256, activation='relu',
                    kernel_regularizer=regularizers.l2(0.001))(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)

    outputs = layers.Dense(len(class_names), activation='softmax')(x)

    model = tf.keras.Model(inputs, outputs)

    return model

model = build_model()

# Optimizador con decaimiento de peso
optimizer = optimizers.AdamW(
    learning_rate=LR_PRETRAIN,
    weight_decay=0.001
)

model.compile(
    optimizer=optimizer,
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'] # Removed 'top_k_categorical_accuracy'
)

# Callbacks mejorados
callbacks = [
    EarlyStopping(
        monitor='val_accuracy',
        patience=15,
        restore_best_weights=True,
        mode='max'
    ),
    ReduceLROnPlateau(
        monitor='val_loss',
        factor=0.2,
        patience=5,
        min_lr=1e-7,
        mode='min'
    ),
    ModelCheckpoint(
        'best_model.keras',
        monitor='val_accuracy',
        save_best_only=True,
        mode='max'
    )
]

# === 11. ENTRENAMIENTO CON WARMUP ===
print("\nüî• Entrenamiento inicial")
history = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_PRETRAIN,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# === 12. FINE-TUNING SELECTIVO ===
print("\nüéØ Fine-tuning avanzado")
base_model = model.layers[1]
base_model.trainable = True

# Descongelar solo las √∫ltimas capas convolucionales
for layer in base_model.layers[:120]:
    layer.trainable = False

model.compile(
    optimizer=optimizers.AdamW(LR_FINETUNE, weight_decay=0.0001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy'] # Removed 'top_k_categorical_accuracy'
)

history_fine = model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=EPOCHS_FINETUNE,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# === 13. GUARDAR MODELO ===
model.save("product_classifier_final.keras")
print("\n‚úÖ Modelo final guardado como 'product_classifier_final.keras'")

# === 14. EVALUACI√ìN DETALLADA ===
def evaluate_model(model, dataset):
    results = model.evaluate(dataset, verbose=0)
    print(f"\nüîπ Loss: {results[0]:.4f}")
    print(f"üîπ Accuracy: {results[1]:.4f}")

    y_true, y_pred = [], []
    for images, labels in dataset.unbatch():
        preds = model.predict(tf.expand_dims(images, axis=0), verbose=0)[0]
        y_true.append(labels.numpy())
        y_pred.append(np.argmax(preds))


    print("\nüìù Reporte de clasificaci√≥n:")
    print(classification_report(y_true, y_pred, target_names=class_names, zero_division=0))

    # Matriz de confusi√≥n para clases con muestras
    present_classes = np.unique(y_true + y_pred)
    present_class_names = [class_names[i] for i in present_classes]

    cm = confusion_matrix(y_true, y_pred, labels=present_classes)
    plt.figure(figsize=(12, 10))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=present_class_names)
    disp.plot(xticks_rotation=90, values_format='d')
    plt.title("Matriz de Confusi√≥n (Clases presentes)")
    plt.tight_layout()
    plt.show()

evaluate_model(model, val_ds)

# === 15. VISUALIZACI√ìN DE RESULTADOS ===
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.title('Accuracy durante el Entrenamiento')
    plt.xlabel('√âpoca')
    plt.ylabel('Accuracy')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.title('Loss durante el Entrenamiento')
    plt.xlabel('√âpoca')
    plt.ylabel('Loss')
    plt.legend()

    plt.tight_layout()
    plt.show()

print("\nüìà Visualizaci√≥n del progreso de entrenamiento:")
plot_training_history(history)
if 'history_fine' in locals():
    plot_training_history(history_fine)

```
